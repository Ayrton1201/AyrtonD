## Jaccard-coëfficient toepassen

De Jaccard-coëfficiënt is een statistische maatstaf die wordt gebruikt om de overeenkomst tussen twee sets te meten. Het is een maat voor de gelijkenis tussen de elementen van de sets.

De Jaccard-coëfficiënt wordt berekend als de verhouding van het aantal elementen dat in beide sets voorkomt tot het totale aantal elementen in de sets. Het wordt uitgedrukt als een waarde tussen 0 en 1, waarbij 0 betekent dat de sets geen gemeenschappelijke elementen hebben en 1 betekent dat de sets volledig identiek zijn.

### Jaccard-coëfficiënt, is een statistiek die wordt gebruikt voor het meten van de gelijkenis en diversiteit van steekproefreeksen.

```python
# De aantallen van de tags optellen
from collections import Counter

tag_counts = Counter(tags)
#print(tag_counts)

tag_counts_df = pd.DataFrame.from_dict(tag_counts, orient='index', columns=['count'])
tag_counts_df.index.name = 'tag'
print(tag_counts_df)
```

```python
# Dictionary aanmaken om de scores van elk recept op te slaan
recipe_scores = {}
# Elk recept langsgaan
for recipe in lekker_recepten["recipe"].to_list():
    # Vraag de df_tag dataframe om de tags voor elk recept die langs komt
    recipe_tags = df_tag.query("recipe=='" + recipe + "'")
    # Elk recept die langs komt start met een score van 0
    score = 0
    # Ga elke tag langs voor het huidige recept
    for tag in recipe_tags["tag"].to_list():
        # Voeg de telling van de huidige tag toe aan de score voor het recept
        score += tag_counts[tag]   
    # Voeg de score van het huidige recept toe aan de recipe_scores dictionary
    recipe_scores[recipe] = score

print(recipe_scores)
```

Stap 4: Nu wil ik een LP model aanmaken met als uitkomst 7 dagen met elke dag 2 recepten. Een recept mag niet vaker dan 1 keer voorkomen. Ik wil dat het aantal kcal per dag uitkomt op 950-1050 kcal en dat de score gemaximaliseerd wordt. Uiteindelijk wil ik dit overzichtelijk maken. Tot slot ook nog 2 plots met de ene plot de kcal per dag over een week en de andere de score op een dag over een week.

```python
import pandas as pd
import matplotlib.pyplot as plt
from pulp import *

# Inladen van de dataframe met 'recipe' 'score' 'value'
result_df

# Aanmaken van het LP probleem
prob = LpProblem('Recipe Planner', LpMaximize)

# Aanmaken van beslissings variabelen
days = range(1, 8)  # 7 dagen in de week
recipes = result_df['recipe'].unique()
recipe_vars = LpVariable.dicts('Recipe', [(day, recipe) for day in days for recipe in recipes], 
                              lowBound=0, cat='Binary')

# Elke dag worden er 2 recepten gekozen en recept mag niet dubbel op een dag
recipe_used_vars = LpVariable.dicts('RecipeUsed', recipes, lowBound=0, cat='Binary')

# Maximale score proberen te krijgen
total_score = lpSum([recipe_vars[(day, recipe)] * result_df.loc[result_df['recipe'] == recipe, 'score'].values[0] 
                     for day in days for recipe in recipes])
prob += total_score

# 2 recepten per dag, geen dubbele recepten per dag
# Totale calorien (value) tussen de 950 en 1050
for day in days:
    prob += lpSum([recipe_vars[(day, recipe)] for recipe in recipes]) == 2
    for recipe in recipes:
        prob += lpSum([recipe_vars[(day, recipe)]]) <= 1
        prob += recipe_vars[(day, recipe)] <= recipe_used_vars[recipe]

for recipe in recipes:
    prob += lpSum([recipe_vars[(day, recipe)] for day in days]) == recipe_used_vars[recipe]

    prob += recipe_used_vars[recipe] <= lpSum([recipe_vars[(day, recipe)] for day in days])

    prob += recipe_used_vars[recipe] <= 1


    prob += recipe_used_vars[recipe] >= lpSum([recipe_vars[(day, recipe)] for day in days]) - 6 * (1 - recipe_used_vars[recipe])

    prob += recipe_used_vars[recipe] >= 0

    prob += recipe_used_vars[recipe] <= 1

    prob += recipe_used_vars[recipe] >= lpSum([recipe_vars[(day, recipe)] for day in days]) - 6

# Waarde beperkingen toevoegen
for day in days:
    prob += lpSum([recipe_vars[(day, recipe)] * result_df.loc[result_df['recipe'] == recipe, 'value'].values[0] for recipe in recipes]) >= 950
    prob += lpSum([recipe_vars[(day, recipe)] * result_df.loc[result_df['recipe'] == recipe, 'value'].values[0] for recipe in recipes]) <= 1050

# Probleem oplossen
prob.solve()
```

```python
# Resultaten printen, alle dagen met 2 recepten, kcal aantal en score.
# Totale calorien en Totale score
for day in days:
    print(f"Day {day}:")
    for recipe in recipes:
        if recipe_vars[(day, recipe)].varValue == 1:
            print(f"  Recipe: {recipe}: score {result_df.loc[result_df['recipe'] == recipe, 'score'].values[0]}, kcal {result_df.loc[result_df['recipe'] == recipe, 'value'].values[0]}")
    print(f"  Total kcal: {lpSum([recipe_vars[(day, recipe)] * result_df.loc[result_df['recipe'] == recipe, 'value'].values[0] for recipe in recipes]).value()}")
    print(f"  Total score: {lpSum([recipe_vars[(day, recipe)] * result_df.loc[result_df['recipe'] == recipe, 'score'].values[0] for recipe in recipes]).value()}")

# Haal de total value en total score voor elke dag eruit om hierna een plot te kunnen maken
values = [lpSum([recipe_vars[(day, recipe)] * result_df.loc[result_df['recipe'] == recipe, 'value'].values[0] for recipe in recipes]).value() for day in days]
scores = [lpSum([recipe_vars[(day, recipe)] * result_df.loc[result_df['recipe'] == recipe, 'score'].values[0] for recipe in recipes]).value() for day in days]
```

```python
# Maak een barplot met de Kcal per dag
fig, ax = plt.subplots()
ax.bar(days, values)
ax.set_xlabel('Day')
ax.set_ylabel('Kcal')
ax.set_title('Total Kcal per Day')
plt.show()
```

```python
# Maak een barplot met de Score per dag
fig, ax = plt.subplots()
ax.bar(days, scores)
ax.set_xlabel('Day')
ax.set_ylabel('Score')
ax.set_title('Total Score per Day')
plt.show()
```
